## 4. Understanding `return` and `pass`

When you write functions, two words you often see are `return` and `pass`.


### 4.1 What does `return` do?

`return` sends a value back from a function.

Example:


````python
def square(x):
    return x * x
````

If you run `square(4)`, the function sends back `16`.

So:
- input: `4`
- output: `16`

`return` is what gives you the answer from the function.


````python
print(square(4))
print(square(7))
````

### 4.2 What does `pass` do?

`pass` means **do nothing for now**.

It is often used as a temporary placeholder while you are still writing your code.


````python
def unfinished_function():
    pass
````

This function is allowed to exist, but it does not do any work yet.

You will see `pass` in starter code later in the notebook.
Your job will be to replace it with real code.


### 4.3 Quick summary

- `return` gives back an answer from a function
- `pass` is a temporary placeholder
- if a function still has `pass`, it is not finished yet

## 6. Absolute error

If:
- $x$ is the **exact value**
- $\hat{x}$ is the **approximate value**

then the **absolute error** is:

$$
\text{absolute error} = |x - \hat{x}|
$$


### Your task

Write a function called `absolute_error(x, x_hat)`.

Your function must:
1. take two inputs: `x` and `x_hat`
2. calculate the absolute error
3. return the answer


### Input and output examples for `absolute_error`

**Example 1**
- Input: `x = 2.735`, `x_hat = 2.70`
- Output: `0.035`

**Example 2**
- Input: `x = 120.0`, `x_hat = 118.5`
- Output: `1.5`

**Example 3**
- Input: `x = 5`, `x_hat = 7`
- Output: `2`


````python
def absolute_error(x, x_hat):
    return abs(x - x_hat)
````

### Test your function

After writing your function, run the next cell.

You should get:
- `0.035`
- `1.5`
- `2`


````python
print(absolute_error(2.735, 2.70))
print(absolute_error(120.0, 118.5))
print(absolute_error(5, 7))
````

## 7. Relative error

The **relative error** is:

$$
\text{relative error} = \frac{|x - \hat{x}|}{|x|}
$$

for $x \neq 0$.


### Your task

Write a function called `relative_error(x, x_hat)`.

Your function must:
1. take two inputs: `x` and `x_hat`
2. calculate the relative error
3. return the answer as a decimal


### Input and output examples for `relative_error`

**Example 1**
- Input: `x = 120.0`, `x_hat = 118.5`
- Output: `0.0125`

**Example 2**
- Input: `x = 0.33`, `x_hat = 0.30`
- Output: about `0.090909...`

**Example 3**
- Input: `x = 2500`, `x_hat = 2450`
- Output: `0.02`


````python
def relative_error(x, x_hat):
    return abs(x - x_hat) / abs(x)
````

### Test your function

After writing your function, run the next cell.

You should get:
- `0.0125`
- about `0.090909...`
- `0.02`


````python
print(relative_error(120.0, 118.5))
print(relative_error(0.33, 0.30))
print(relative_error(2500, 2450))
````

## 10. While loops

A **while loop** repeats code while a condition is true.

### General form
```python
while condition:
    # repeated code
```

A while loop is useful when you want to keep repeating something until you reach a target.


## 11. Example: a while loop with absolute error

Read the example below.

It:
1. starts with `exact_value = 10`
2. starts with `approx_value = 6`
3. increases `approx_value` by `1`
4. stops when the absolute error is less than `1`


````python
exact_value = 10
approx_value = 6

while absolute_error(exact_value, approx_value) >= 1:
    print("Current approximation:", approx_value)
    print("Current absolute error:", absolute_error(exact_value, approx_value))
    approx_value = approx_value + 1

print("Final approximation:", approx_value)
print("Final absolute error:", absolute_error(exact_value, approx_value))
````

## 15. A pre-written approximation function

Sometimes the approximation method is already written for you.

In the next example, the function below approximates $\sqrt{2}$.


````python
def approximate_sqrt2(steps):
    guess = 1.0
    count = 0

    while count < steps:
        guess = 0.5 * (guess + 2 / guess)
        count = count + 1

    return guess
````

### Input and output examples for `approximate_sqrt2`

**Example 1**
- Input: `steps = 1`
- Output: an approximation to `sqrt(2)`

**Example 2**
- Input: `steps = 3`
- Output: a better approximation to `sqrt(2)`

**Example 3**
- Input: `steps = 5`
- Output: an even better approximation to `sqrt(2)`


````python
print(approximate_sqrt2(1))
print(approximate_sqrt2(3))
print(approximate_sqrt2(5))
print("Exact value:", math.sqrt(2))
````
